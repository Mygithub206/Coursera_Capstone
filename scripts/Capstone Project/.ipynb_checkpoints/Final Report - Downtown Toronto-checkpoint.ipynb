##### Foursquare API:  
This project would use Four-square API as its prime data gathering source as it has a database of millions of places, especially their places API which provides the ability to perform location search, location sharing and details about a business.

The data retrieved from Foursquare contained information of venues within a specified distance of the longitude and latitude of the postcodes. The information obtained per venue as follows:

    1. Neighborhood
    2. Neighborhood Latitude
    3. Neighborhood Longitude
    4. Venue
    5. Name of the venue e.g. the name of a store or restaurant
    6. Venue Latitude
    7. Venue Longitude
    8. Venue Category
    
##### Toronto Public Safety Data API:  
This Robbery dataset is a subset of the Major Crime Indicators (MCI) from 2014 to 2019. Shared By: TorontoPoliceService  

The data retrieved from Toronto Police Service contained information of rebberies with a specified distance if the longitude and latitude. The information obtained per robbery as follows:

    Division, event_unique_id, Hood_ID, MCI, Neighbourhood, occurrencedate, Date or Time, occurrenceday,  
    occurrencedayofweek, occurrencedayofyear, occurrencehour, occurrencemonth, occurrenceyear, offence, premisetype,  
    reporteddate, reportedday, reporteddayofweek, reporteddayofyear, reportedhour, reportedmonth, reportedyear,  
    ucr_code, ucr_ext
